```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(data)
```

```python
x_scaled
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
X_pca_df = pd.DataFrame(x_pca, columns=['PC1', 'PC2'])
print(X_pca_df.head())
```

```python
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained:", sum(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c='blue', edgecolor='k', s=50)
plt.title('PCA - Dimensionality Reduction')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(7,5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, 'o-', color='green')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.grid(True)
plt.axhline(y=0.90, color='r', linestyle='--')  # e.g., 90% threshold
plt.show()
```

```python
for i, var in enumerate(cumulative_variance):
    print(f"PC {i+1}: Cumulative Variance = {var:.4f}")
```

```python
pca_columns = [f'PC{i+1}' for i in range(2)]
X_pca_df = pd.DataFrame(x_pca, columns=pca_columns)
```

```python
from sklearn.cluster import KMeans
```

```python
k = 3  # Example: 3 clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(x_scaled)

# Get cluster labels for each sample
labels = kmeans.labels_
```

```python
data['Cluster'] = labels
print(data.head())
```

```python
X_pca_df['Cluster'] = data['Cluster']
plt.figure(figsize=(7,5))
sns.scatterplot(x='PC1', y='PC2',hue='Cluster', data=X_pca_df,palette='Set1',s=50,alpha=0.8)
plt.title('K-Means Clustering on PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
labels_original = kmeans.labels_
```

```python
comparison_df = pd.DataFrame({
    'Original_Cluster': labels_original,
    'PCA_Cluster': labels
})

print(pd.crosstab(comparison_df['Original_Cluster'], comparison_df['PCA_Cluster']))
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_score = silhouette_score(x_scaled, labels)
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(data)

# Apply KMeans clustering on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_pca, cmap='viridis', s=50)
plt.scatter(kmeans_pca.cluster_centers_[:,0], kmeans_pca.cluster_centers_[:,1],
            color='red', marker='X', s=200, label='Centroids')
plt.title('Clustering on PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# Original KMeans clustering on full data (assuming 3 clusters)
kmeans_original = KMeans(n_clusters=3, random_state=42)
labels_original = kmeans_original.fit_predict(x_scaled)

# Plot side by side
fig, axes = plt.subplots(1, 2, figsize=(14,6))

axes[0].scatter(x_scaled[:,0], x_scaled[:,1], c=labels_original, cmap='viridis', s=50)
axes[0].set_title('Original Data Clustering')

axes[1].scatter(X_pca[:,0], X_pca[:,1], c=labels_pca, cmap='viridis', s=50)
axes[1].set_title('PCA-Transformed Data Clustering')

plt.show()
```

```python
In cluster assignments the original dataset have based on all features and PCA transformed Dataset have based on principal components.
Silhouette Score in original dataset is 0.62 and PCA transformed dataset is 0.58
```

```python
. Similarities and Differences

Similarities:

Overall cluster structure is often preserved if the top principal components capture most of the variance.

Major groups and patterns can still be detected in PCA-transformed data.

Differences:

Some fine-grained distinctions may be lost after PCA (especially if only a few components are retained).

PCA may slightly alter cluster centroids and assignments due to variance reduction.

Clusters in PCA space are easier to visualize and interpret in 2D or 3D plots.

3. Impact of dimensionality reduction on clustering performance

Advantages:

Reduces noise and redundant features, improving clustering stability.

Faster computation for high-dimensional data.

Easier visualization of clusters in reduced dimensions.

Potential drawbacks:

Loss of information may slightly reduce cluster quality (lower silhouette or higher DBI).

Rare patterns in low-variance features might be missed.
```

```python
Key Findings and Insights

Clustering on Original Data:

KMeans successfully identified natural groupings in the dataset.

Silhouette Score and Daviesâ€“Bouldin Index indicated reasonably well-separated and compact clusters.

Visualization of high-dimensional clusters was challenging without dimensionality reduction.

Clustering on PCA-Transformed Data:

PCA reduced dimensionality while retaining most of the variance.

Clustering on the reduced dataset produced results similar to the original data, indicating that PCA preserved the underlying cluster structure.

Silhouette Scores and DBI were slightly lower/higher, reflecting minor loss of information.

Visualization in 2D/3D became easier, making cluster interpretation more intuitive.

Comparison:

Major cluster patterns were consistent between the original and PCA-transformed datasets.

PCA slightly altered some cluster boundaries but improved computational efficiency and visualization.

2. Practical Implications

PCA:

Useful for high-dimensional datasets to reduce noise and redundancy.

Facilitates faster computation and clear visualization of clusters.

Helps in identifying the most informative features contributing to variance.

Clustering:

Effective for unsupervised pattern recognition and segmenting data without prior labels.

Provides insights into underlying structure of datasets, which can inform marketing, risk assessment, or scientific analysis.

Performance can be influenced by dimensionality and irrelevant features.

Combined Use (PCA + Clustering):

Balances dimensionality reduction and cluster detection.

Suitable for exploratory data analysis, especially when the dataset has many features.

3. Recommendations

Use clustering on original data when:

Dataset has a manageable number of dimensions.

All features are meaningful and should be preserved.

High fidelity in cluster assignments is critical.

Use PCA before clustering when:

Dataset is high-dimensional with redundant or correlated features.

Visualization of clusters in 2D/3D is required.

Computational efficiency is a concern.

Noise reduction is beneficial to improve clustering stability.

Key takeaway:

Combining PCA with clustering is a practical approach for exploratory data analysis, offering a trade-off between interpretability, performance, and computational efficiency.

Direct clustering works best when feature integrity is more important than visualization or speed.
```
