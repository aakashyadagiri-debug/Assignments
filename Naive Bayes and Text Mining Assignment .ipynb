```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab') # Download the missing resource

# Sample text
text = "Hello! This is an example sentence, showing how to clean text data."

# 1. Convert to lowercase
text = text.lower()

# 2. Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# 3. Tokenize the text
tokens = word_tokenize(text)

# 4. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Original Text:", text)
print("Tokens:", tokens)
print("After Removing Stopwords:", filtered_tokens)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample preprocessed text data (list of documents)
corpus = [
    "hello example sentence showing clean text data",
    "machine learning is fun and powerful",
    "naive bayes is a simple and effective algorithm"
]

# 1. Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# 2. Fit and transform the text data
X_tfidf = vectorizer.fit_transform(corpus)

# 3. View feature names and transformed data
print("Feature Names:", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n", X_tfidf.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

# Drop irrelevant columns from X
X = data[['Data']]
Y = data['Labels']

# Drop rows with NaN values in 'Data' or 'Labels'
data_cleaned = data.dropna(subset=['Data', 'Labels'])
X_cleaned = data_cleaned['Data']
Y_cleaned = data_cleaned['Labels']

# Split the cleaned data
x_train, x_test, y_train, y_test = train_test_split(X_cleaned, Y_cleaned, test_size=0.2, random_state=50)

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training data
x_train_tfidf = vectorizer.fit_transform(x_train)

# Transform the testing data
x_test_tfidf = vectorizer.transform(x_test)

# Train the Multinomial Naive Bayes model
model = MultinomialNB()
model_lb = model.fit(x_train_tfidf, y_train)
```

```python
y_pred=model.predict(x_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred,average='weighted')
rec=recall_score(y_test,y_pred,average='weighted')
f1=f1_score(y_test,y_pred,average='weighted')
```

```python
print(f"accuracy: {acc}")
print(f"precision: {prec}")
print(f"recall: {rec}")
print(f"f1_score: {f1}")
```

```python
from textblob import TextBlob
```

```python
data = {
    'blog_post': [
        "I absolutely love working on machine learning projects!",
        "This website is terrible and slow.",
        "The new update is okay, but could be better."
    ]
}

df = pd.DataFrame(data)
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity  # Range: -1 (negative) → +1 (positive)
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['blog_post'].apply(get_sentiment)
print(df)
```

```python
Performance of the Model

When you build a sentiment classification model (e.g., using Naive Bayes, Logistic Regression, or any ML classifier), the performance can be evaluated using the following metrics:

Common Metrics

Accuracy

Fraction of correct predictions out of all predictions.

Good for balanced datasets.

Precision

Out of all instances predicted as a class (e.g., Positive), how many were actually that class.

Important when false positives are costly.

Recall (Sensitivity)

Out of all actual instances of a class, how many did the model correctly identify.

Important when missing a positive instance is costly.

F1-Score

Harmonic mean of precision and recall.

Balances precision and recall, especially for imbalanced classes.

Confusion Matrix

Shows the true positives, true negatives, false positives, and false negatives.

Helpful for visual inspection of misclassifications.

Challenges Encountered During Classification
a) Imbalanced Data

Often, blogs might have more Positive posts than Negative or Neutral.

Models tend to bias toward majority classes.

Solution: Use techniques like SMOTE, class weights, or stratified splitting.

b) Ambiguous or Mixed Sentiments

Some blog posts contain both positive and negative statements, e.g.,
“The article was informative but poorly written.”

Models may misclassify these as Neutral or Positive incorrectly.

Solution: Consider fine-grained sentiment analysis or use contextual NLP models.

c) Short Texts or Noisy Data

Short blogs or informal text may lack context, making polarity detection harder.

Solution: Preprocessing like stopword removal, lemmatization, and text cleaning.

d) Sarcasm and Irony

Sentences like “Oh great, another useless tutorial.” can confuse models.

Traditional ML struggles with sarcasm.

Solution: Advanced models like BERT or RoBERTa handle context better.

e) Feature Representation

Bag-of-Words or TF-IDF may lose context of words.

Word embeddings (Word2Vec, GloVe) or transformer-based models improve accuracy.
```

```python
A majority of positive posts indicates that the content is generally well-received, engaging, or informative.

Negative posts highlight areas where the blog content may have confused, disappointed, or frustrated readers.

Neutral posts suggest content that is informative but does not evoke strong emotions.

Certain categories (e.g., Health) may have more positive sentiments, reflecting that readers value informative or helpful content in that area.

Categories like Finance may have mixed or more negative sentiments, possibly due to complex topics, differing opinions, or controversial subjects.

Neutral sentiment in some categories suggests content is factual, informative, but less emotionally engaging.

3. Insights from Sentiment Analysis

Content Strengths

Positive posts often highlight clear, useful, or engaging content.

Consistent positive sentiment can guide the blog creators to focus on these successful styles or topics.

Areas for Improvement

Negative posts point out gaps, unclear explanations, or reader dissatisfaction.

Blogs with repeated negative feedback might need revisions, better clarity, or improved formatting.

Reader Engagement

The sentiment analysis gives an indirect measure of how readers feel about the content.

Categories with high positive sentiment may attract more engagement or shares.

Categories with high negative sentiment may require editorial review or deeper research.

Neutral Posts

Neutral posts often represent informative, unbiased content.

While they do not evoke strong emotion, they can still serve as educational or reference material.

4. Recommendations

Emphasize positive patterns: replicate writing style, tone, or topics that generate positive sentiment.

Address negative feedback: revise or clarify topics that readers respond negatively to.

Monitor neutral posts: ensure they are informative and enhance value to readers.

Segment content strategies by category: focus on strengths and weaknesses in each category.
```
