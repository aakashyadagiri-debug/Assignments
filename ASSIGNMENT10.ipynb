```python
# Check for any missing values
df.isnull().sum()
```

```python
# Remove rows that contain any missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# transactions to list of items
transactions = df.iloc[:, 0].apply(lambda x: x.split(',')).tolist()
# first 5 transactions
for i, transaction in enumerate(transactions[:5]):
    print(f"Transaction {i+1}:", transaction)
```

```python
#ASSOCIATION RULE MINING
```

```python
#ONE HOT ENCODING
# One-hot encoding is used to convert transaction data into a binary format required for association rule mining algorithms like Apriori.
```

```python
#install mlxtend
!pip install mlxtend
```

```python
#importing libraries
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori , association_rules
# Initialize encoder and fit-transform the transactions
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
# Convert the array to a DataFrame
basket_df = pd.DataFrame(te_array, columns=te.columns_)
basket_df
```

```python
#APRIORI ALGORITHM
# Apriori algorithm is used to identify frequent itemsets and generate association rules in transactional datasets for market basket analysis.
```

```python
# Get frequent itemsets with min support
frequent_itemsets = apriori(basket_df, min_support=0.01, use_colnames=True)
# View top itemsets
frequent_itemsets.sort_values(by="support", ascending=False).head()
```

```python
# Generate the rules from frequent_itemsets
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
# Filter rules with meaningful thresholds
filtered_rules = rules[(rules['confidence'] > 0.3) & (rules['lift'] > 3)]
# View top rules
filtered_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

```python
 Analysis and Interpretation
1. Analyse the generated rules:
The generated association rules revealed strong product pairings such as Green Tea â†’ Honey and Milk â†’ Bread.
High confidence and lift values indicate these combinations are frequently and consistently purchased together.
Lift values greater than 3 show that the associations are significant and not due to chance.
Such rules can help identify product bundling opportunities.

2. Interpret the results and provide insights:
The rules suggest that customers often buy complementary products, especially health-related and daily-use items.
For example, buyers of mineral water often purchase salmon, indicating a health-conscious segment.
Patterns show consistent consumer behavior, which can guide personalized offers and store layout.
This analysis supports better marketing and sales strategies.
```

```python
Interview Questions

1. What is lift and why is it important in Association rules?
-->Lift measures how much more likely item B is purchased when item A is bought, compared to random chance.
It is calculated as:
Lift(ð´â†’ðµ)=Confidence(ð´â†’ðµ)/Support(ðµ)
-->Lift > 1 indicates a meaningful and strong relationship between A and B.

2. What is support and confidence? How do you calculate them?

>Support shows how frequently an item or itemset occurs:
-->Support(A) = (Transactions containing A) / (Total transactions)
>Confidence shows the likelihood of item B being purchased when A is bought:
-->Confidence(A â†’ B) = (Transactions with A & B) / (Transactions with A)
 
3. What are some limitations or challenges of Association rules mining?

*It can generate too many rules, many of which may be trivial or irrelevant.
*Rare but important patterns might be missed due to low support.
*The technique doesn't consider the sequence or timing of transactions.
*Works only with categorical or binary (one-hot encoded) data formats.

```
